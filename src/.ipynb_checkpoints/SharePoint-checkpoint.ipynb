## delete_version_history(base_url, file_url, file_name, delete=False, verbose=False, last_remaining=None):

    * base_url:
        base_url
        
    * file_url:
        file_url
        
    * file_name:
        file name
        
    * delete:
        setting delete to True will perform the delete action, False won't
    
    * verbose:
        if verbose, you get additional information regarding file and its version history.
    
    * last_remaining: 
        means number of version history it is wished NOT to be deleted.
        last_remaining = 5, means 5 newest version history will NOT be deleted.

    

```python
def delete_version_history(base_url, file_url, file_name, delete=False, verbose=False, last_remaining=10):
    
    client_credentials = ClientCredential(client_id=app_principal['client_id'],client_secret=app_principal['client_secret'])
    client = ClientContext("https://eaenergianalyse.sharepoint.com/" + base_url).with_credentials(client_credentials)
    file = client.web.get_file_by_server_relative_path(file_url + file_name).get().execute_query()
    file_ver = client.web.get_file_by_server_relative_path(file_url + file_name).versions.expand(['Created']).get().execute_query()
    
    if verbose:
        print("File Name: ", file_name)
        print("Location: ", file_url)
        print("Time Created: ", str(pd.to_datetime(file.properties["TimeCreated"])))
        print("Last Time Modified: ",str(pd.to_datetime(file.properties["TimeLastModified"])))
        print("Major Ver.: ", file.properties["MajorVersion"])
        print("Sorted by oldest")
        print(f"\n\tindex:\tDate created:  \t\t\tSize:")
    total_len = int()
    for cnt, i in enumerate(range(len(file_ver.__dict__['_data']))):
        total_len += int(file_ver.__dict__['_data'][i].__dict__["_properties"]["Length"])

        date = str(pd.to_datetime(file_ver.__dict__['_data'][i].__dict__["_properties"]["Created"]))
        length = round(int(file_ver.__dict__['_data'][i].__dict__["_properties"]["Length"]) / 1000000,2)
        if verbose:
            print(f"\t{cnt+1}: \t{date} \t{length} MB ")
    if verbose:
        print("Total ver. history size: ", round(total_len / 1000000,2), "MB\n")
    
    _len = len(file_ver.__dict__['_data'])
    n=last_remaining
    if last_remaining == None or last_remaining == 0:        
        if delete:
            for _ in range(len(file_ver.__dict__['_data'])):
                file_ver.__dict__['_data'][0].delete_object().execute_query()
            if verbose:
                print("Version history has been permenantly deleted!")
                
    elif last_remaining >= 1:
        if delete:
            for c,i in enumerate(range((-n+1) + _len)):
                file_ver.__dict__['_data'][0].delete_object().execute_query()
            if verbose:
                print(f"Except {last_remaining}, the rest has been permenantly deleted!")
    return  
```

```python
class view_contents():
    
    def __init__(self, base_url, target_url):
        client_credentials = ClientCredential(client_id=app_principal['client_id'],client_secret=app_principal['client_secret'])
        client = ClientContext(site_url + base_url).with_credentials(client_credentials)
        
        self.base_url = base_url
        self.folder = client.web.get_folder_by_server_relative_url(target_url)
        
        self.directory = self.folder.folders
        client.load(self.directory).execute_query()
        
        self.sub_folders = self.folder.files 
        client.load(self.sub_folders).execute_query()
        
        # these folders are ignored
        self.ignore = ['_catalogs','m','_cts','Lists','_private','_vti_pvt','SitePages','Delingslinks','images','Forms','SiteAssets']

    def _folders(self):
        
        folders = []
        folder_props = {}
        for d in self.directory._data:
            if d.properties["Name"] not in self.ignore:
                folders.append(d.properties["Name"])

        return folders 
            
    def _files(self, show_files = False):  
        
        files = []
        file_props = {}
        for s_folder in self.sub_folders:
            files.append(s_folder.properties["Name"])
        
        if show_files:
            print(f'files in {target_url }:\n')
            print(f"\tindex\tfiles:")
            for cnt,item in enumerate(range(len(files))):
                print(f'\t{cnt}\t{files[item]}')
        
        all_values = {
            
            "files":files,
            "target_url":target_url,
            "base_url":self.base_url
        }
        
        return all_values
```

```python
def general_info(json_data):

    title, id_, url, sru, rb, desc, created, lim, lium = [],[],[],[],[],[],[],[],[]

    for i in range(len(json_data['d']['results'])):
        title.append(json_data['d']['results'][i]['Title'])
        id_.append(json_data['d']['results'][i]['Id'])
        url.append(json_data['d']['results'][i]['Url'])
        sru.append(json_data['d']['results'][i]['ServerRelativeUrl'])
        rb.append(json_data['d']['results'][i]['RecycleBinEnabled'])
        desc.append(json_data['d']['results'][i]['Description'])
        created.append(json_data['d']['results'][i]['Created']) 
        lim.append(json_data['d']['results'][i]['LastItemModifiedDate'])
        lium.append(json_data['d']['results'][i]['LastItemUserModifiedDate']) 


    info = {

        "Title": title,
        "Id" : id_,
        "Url": url,
        "ServerRelativeUrl": sru,
        "RecycleBin" : rb,
        "Desc" : desc,
        "Created" : created, 
        "LastItemModified":  lim,
        "LastItemUserModified" :  lium,    
    }

    return pd.DataFrame(info, columns= info.keys())
```

# Clean the Recycle Bin
### Notice that cleaning the recycle bin will permenantly delete files from recycle bin

```python
def clean_recycle_bin(delete = False):
    """
    https://www.sharepointdiary.com/2016/06/sharepoint-online-powershell-to-empty-recycle-bin.html
    """
    print("Empty recycle bins are excluded")
    if not delete:
        print("Delete is set to False. You can veiw the recycle bin.")
    else:
        print("Deleted items:")
    
    client_credentials = ClientCredential(client_id=app_principal['client_id'],client_secret=app_principal['client_secret'])

    for i in range(len(urls)):

        client = ClientContext(urls[i]).with_credentials(client_credentials)

        Site = client.site
        RecycleBinItems = Site.recycle_bin
        client.load(Site)
        try: 
            
            client.load(RecycleBinItems).execute_query()
            if len(RecycleBinItems.__dict__['_data']) != 0:
                
                if delete:
                    RecycleBinItems.delete_all().execute_query() 
                    print(f"\t{site_titles[i]}: {len(RecycleBinItems.__dict__['_data'])} file/s")
                if not delete:
                    print(f"\t{site_titles[i]}: {len(RecycleBinItems.__dict__['_data'])} file/s")                 
                    
        except Exception:
            #print(f"\t>>>{site_titles[i]}: Access to the recycle bin is denied!<<<")
            pass
    if delete:
        print("All recycle bins are permanently deleted!")
    if not delete:
        print("All non-empty recycle bins can be viewed.")
```

### If delete = False, you can view all the Recycle Bin's content.

```python
def version_history(base_url, file_url, file_name):
    

    client_credentials = ClientCredential(client_id=app_principal['client_id'],client_secret=app_principal['client_secret'])
    client = ClientContext("https://eaenergianalyse.sharepoint.com/" + base_url).with_credentials(client_credentials)
    file = client.web.get_file_by_server_relative_path(file_url + file_name).get().execute_query()
    file_ver = client.web.get_file_by_server_relative_path(file_url + file_name).versions.expand(['Created']).get().execute_query()
    
    print("File Name: ", file_name)
    print("Location: ", file_url)
    print("Time Created: ", str(pd.to_datetime(file.properties["TimeCreated"])))
    print("Last Time Modified: ",str(pd.to_datetime(file.properties["TimeLastModified"])))
    print("Major Ver.: ", file.properties["MajorVersion"])
    print("Sorted by oldest")
    print(f"\n\tindex:\tDate created:  \t\t\tSize:")
    total_len = int()
    for cnt, i in enumerate(range(len(file_ver.__dict__['_data']))):
        total_len += int(file_ver.__dict__['_data'][i].__dict__["_properties"]["Length"])

        date = str(pd.to_datetime(file_ver.__dict__['_data'][i].__dict__["_properties"]["Created"]))
        length = round(int(file_ver.__dict__['_data'][i].__dict__["_properties"]["Length"]) / 1000000,2)
        print(f"\t{cnt+1}: \t{date} \t{length} MB ")
    print("Total ver. history size: ", round(total_len / 1000000,2), "MB\n")
    
    _len = len(file_ver.__dict__['_data'])
    print(_len)
                
    
    return 
```
