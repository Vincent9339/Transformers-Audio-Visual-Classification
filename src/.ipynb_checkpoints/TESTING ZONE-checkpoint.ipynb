```python
import os, cv2, torch, timm, warnings, librosa, torchaudio, math
import numpy as np
import torch.nn.functional as F
from torchsummary import summary

from timm.models.layers import trunc_normal_
from timm.models.vision_transformer import Block, PatchEmbed

warnings.filterwarnings('ignore')
from PIL import Image
import matplotlib.pyplot as plt
from torchvision import transforms as T
from scipy.io.wavfile import read
from scipy.io import wavfile
from scipy.signal import spectrogram
```

```python
class utils:
    def __init__(self,sample_rate=int, window_size=int, hop_size=int):
        self.sample_rate = sample_rate
        self.window_size = window_size
        self.hop_size = hop_size
        
    def extract_img_audio(self,loc, output_jpg):
        for filename in os.listdir():
            if filename.endwith('.wav'):
                sample_rate, audio_data = wavefile.read(filename)
                f, t, Sxx = spectrogram(audio_data, fs=sample_rate)
                plt.pcolormesh(t, f, 10 * np.log10(Sxx))  
                plt.savefig(output_jpg)
        
    def video_to_wav(self,empty_curr_loc=False):
        
        for filename in os.listdir():
            if filename.endswith('.mp4'):
                if empty_curr_loc:
                    os.system('rm *.wav')
            os.system('ffmpeg -n -i {} -hide_banner -loglevel error -acodec pcm_s16le -ac 1 -ar 16000 {}.wav'.format(filename, filename.split('.')[0]))

    def load_spectrogram(self,wav_file, sample_rate=16000, n_fft=2048, hop_length=512, window='hann', n_mels=128):

        raw_audio, sample_rate = librosa.load(wav_file)
        S = librosa.feature.melspectrogram(y=raw_audio, sr=sample_rate, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels)
        S_dB = librosa.power_to_db(S, ref=np.max)  
        return S_dB,sample_rate
    
    def display_spectrogram(self, audio_data, sample_rate):
        
        plt.figure(figsize=(20, 5))
        librosa.display.specshow(audio_data, x_axis='time', y_axis='mel', sr=sample_rate, hop_length=512)
        plt.colorbar(format='%+2.0f dB')
        plt.title('Mel-frequency spectrogram')
        plt.show()
        
    def np2image(self,array):
        plt.figure(figsize=(20, 5))
        plt.imshow(array, cmap='viridis')
        plt.axis('off')
        plt.show()
        
    def load_image(self,loc):
        
        from IPython.display import Image
        return Image(filename=loc)
    
    def load_audio(self,loc):
        
        audio, sample_rate = librosa.load(loc)
        return audio,sample_rate
    
    def load_video(self,loc):
        frames = []
        cap = cv2.VideoCapture(loc)
        ret = True
        while ret:
            ret, img = cap.read()
            if ret:
                frames.append(img)
        video = np.stack(frames, axis=0)
        return video
    
    def play_audio(self,loc):
        
        from IPython.display import Audio
        raw_audio, sample_rate = librosa.load(loc)
        
        return Audio(data=raw_audio, rate=sample_rate)
    
    def image_resize(self,w:int, h:int,image_loc,display=False):

        w = w
        h = h

        img = Image.open(image_loc)
        wpercent = (w / float(img.size[0]))
        img = img.resize((w, h), Image.Resampling.LANCZOS)
        if display:
            return img
        else:
            return np.array(img)
```

```python
imagenet_mean = np.array([0.485, 0.456, 0.406])
imagenet_std = np.array([0.229, 0.224, 0.225])
img_size=224
patch_size=16
in_chans=3
embed_dim=1024

def load_image(img_loc):
    
    img = Image.open(img_loc)
    img = img.resize((img_size, img_size))
    img = np.array(img) / 255.
    img = img - imagenet_mean
    img = img / imagenet_std
    img = torch.Tensor(img)
    return img

def show_image(image,title=''):
    if image.shape[0] == 3:
        image = torch.einsum('chw->hwc', image)
    plt.rcParams['figure.figsize'] = [7, 7]
    imagenet_mean = np.array([0.485, 0.456, 0.406])
    imagenet_std = np.array([0.229, 0.224, 0.225])
    plt.imshow(torch.clip((image * imagenet_std + imagenet_mean) * 255, 0, 255).int())
    plt.title(title, fontsize=16)
    return

def patchify(imgs):
    """
    imgs: (N, 3, H, W)
    x: (N, L, patch_size**2 *3) 
    N=number of images
    L=number of patches
    patch size=16
    """
    patch_embed = PatchEmbed(img_size, patch_size, in_chans, embed_dim)
    num_patches = patch_embed.num_patches
    p = patch_embed.patch_size[0] #
    assert imgs.shape[2] == imgs.shape[3] and imgs.shape[2] % p == 0

    h = w = imgs.shape[2] // p
    x = imgs.reshape(shape=(imgs.shape[0], 3, h, p, w, p))
    x = torch.einsum('nchpwq->nhwpqc', x)
    x = x.reshape(shape=(imgs.shape[0], h * w, p**2 * 3))
    return x

def random_masking(x, mask_ratio):

    N, L, D = x.shape  # batch, length, dim
    len_keep = int(L * (1 - mask_ratio))

    noise = torch.rand(N, L, device=x.device)  # noise in [0, 1]

    ids_shuffle = torch.argsort(noise, dim=1)  # ascend: small is keep, large is remove
    ids_restore = torch.argsort(ids_shuffle, dim=1)

    ids_keep = ids_shuffle[:, :len_keep]
    x_masked = torch.gather(x, dim=1, index=ids_keep.unsqueeze(-1).repeat(1, 1, D))

    mask = torch.ones([N, L], device=x.device)
    mask[:, :len_keep] = 0
    mask = torch.gather(mask, dim=1, index=ids_restore)

    return x_masked, mask, ids_restore

def return_masked(input_file):
    
    x_masked = input_file
    y=list()
    for i in range(mask.shape[1]):
        if mask[0][i].item()==1.0:
            y.append(torch.zeros(x_masked.shape[2]))
        if mask[0][i].item()==0.0:
            y.append(torch.flatten(x_masked[:,ids_restore[0][i].item():ids_restore[0][i].item()+1]))

    x = torch.stack(y,dim=0).unsqueeze(0)
    return x

def unpatchify(x):
    patch_embed = PatchEmbed(img_size, patch_size, in_chans, embed_dim)
    num_patches = patch_embed.num_patches
    p = patch_embed.patch_size[0]
    h = w = int(x.shape[1]**.5)
    assert h * w == x.shape[1]

    x = x.reshape(shape=(x.shape[0], h, w, p, p, 3))
    x = torch.einsum('nhwpqc->nchpwq', x)
    img = x.reshape(shape=(x.shape[0], 3, h * p, h * p))
    return img
```

```python
def make_features(wav_name, mel_bins, target_length=1024):
    waveform, sr = torchaudio.load(wav_name)
    assert sr == 16000, 'input audio sampling rate must be 16kHz'

    fbank = torchaudio.compliance.kaldi.fbank(
        waveform, htk_compat=True, sample_frequency=sr, use_energy=False,
        window_type='hanning', num_mel_bins=mel_bins, dither=0.0, frame_shift=10)

    n_frames = fbank.shape[0]

    p = target_length - n_frames
    if p > 0:
        m = torch.nn.ZeroPad2d((0, 0, 0, p))
        fbank = m(fbank)
    elif p < 0:
        fbank = fbank[0:target_length, :]

    fbank = (fbank - (-4.2677393)) / (4.5689974 * 2)
    print(type(fbank))
    return fbank

audio = make_features('test.wav', mel_bins=128).unsqueeze(0)
```

### Video Encoder

```python
class v_Encoder(torch.nn.Module):
    def __init__(self,
                 image_size=224,
                 patch_size=16,
                 emb_dim=1024, 
                 num_layer=12,
                 num_head=16,
                 mlp_ratio=4,
                 ) -> None:
        super().__init__()
        self.patch_embed = PatchEmbed(img_size, patch_size, in_chans, embed_dim)
        num_patches = self.patch_embed.num_patches
        self.cls_token = torch.nn.Parameter(torch.zeros(1, 1, emb_dim))
        self.pos_embed = self.sincos_pos_embed()
        x = img.unsqueeze(0)
        x = torch.einsum('nhwc->nchw', x)
        x = patchify(x)
        self.x_masked, _, _ = random_masking(x,mask_ratio=.75)
        self.shuffle = self.x_masked

        self.input = torch.nn.Conv2d(3, emb_dim, patch_size, patch_size)
        self.blocks = nn.ModuleList([
            Block(embed_dim, num_head, mlp_ratio, qkv_bias=True, norm_layer=torch.nn.LayerNorm)
            for i in range(num_layer)])
        self.transformer = torch.nn.Sequential(*[Block(emb_dim, num_head) for _ in range(num_layer)])

        self.norm = torch.nn.LayerNorm(emb_dim)

        self.init_weight()

        
    def patchify(self,imgs):
        """
        imgs: (N, 3, H, W)
        x: (N, L, patch_size**2 *3) 
        N=number of images
        L=number of patches
        patch size=16
        """
        patch_embed = PatchEmbed(img_size, patch_size, in_chans, embed_dim)
        num_patches = patch_embed.num_patches
        p = patch_embed.patch_size[0] #
        assert imgs.shape[2] == imgs.shape[3] and imgs.shape[2] % p == 0

        h = w = imgs.shape[2] // p
        x = imgs.reshape(shape=(imgs.shape[0], 3, h, p, w, p))
        x = torch.einsum('nchpwq->nhwpqc', x)
        x = x.reshape(shape=(imgs.shape[0], h * w, p**2 * 3))
        return x

    def sincos_pos_embed(self,max_sequence_length = 197, embedding_dim=1024):

        position = torch.arange(0, max_sequence_length).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, embedding_dim, 2) * -(math.log(10000.0) / embedding_dim))
        position_encoding = torch.zeros(1, max_sequence_length, embedding_dim)
        position_encoding[0, :, 0::2] = torch.sin(position * div_term)
        position_encoding[0, :, 1::2] = torch.cos(position * div_term)

        return position_encoding
    def random_masking(self,x, mask_ratio):

        N, L, D = x.shape  # batch, length, dim
        len_keep = int(L * (1 - mask_ratio))

        noise = torch.rand(N, L, device=x.device)  # noise in [0, 1]

        ids_shuffle = torch.argsort(noise, dim=1)  # ascend: small is keep, large is remove
        ids_restore = torch.argsort(ids_shuffle, dim=1)

        ids_keep = ids_shuffle[:, :len_keep]
        x_masked = torch.gather(x, dim=1, index=ids_keep.unsqueeze(-1).repeat(1, 1, D))

        mask = torch.ones([N, L], device=x.device)
        mask[:, :len_keep] = 0
        mask = torch.gather(mask, dim=1, index=ids_restore)

        return x_masked, mask, ids_restore

    def init_weight(self):
        trunc_normal_(self.cls_token, std=.02)
        trunc_normal_(self.pos_embed, std=.02)

    def forward(self, img):
        x = img.unsqueeze(0)
        x = torch.einsum('nhwc->nchw', x)
        x = self.patch_embed(x)
        x = x + self.pos_embed[:, 1:, :]
        x, mask, ids_restore = self.random_masking(x,0.75)
        cls_token = self.cls_token + self.pos_embed[:, :1, :]
        cls_tokens = cls_token.expand(x.shape[0], -1, -1)
        x = torch.cat((cls_tokens, x), dim=1)
        for blk in self.blocks:
            x = blk(x)
        x = self.norm(x)
        
        return x, mask, ids_restore

encoder = v_Encoder()
x, mask, ids_restore = encoder(img)
```

### Video Decoder

```python
class v_Decoder(torch.nn.Module):
    def __init__(self,
                 image_size=224,
                 patch_size=16,
                 emb_dim=1024,
                 dec_emb_dim=512, 
                 num_layer=8,
                 num_head=16,
                 in_chans=3,
                 mlp_ratio=4,
                 ) -> None:
        super().__init__()
        self.patch_embed = PatchEmbed(img_size, patch_size, in_chans, embed_dim)
        num_patches = self.patch_embed.num_patches
        self.cls_token = torch.nn.Parameter(torch.zeros(1, 1, emb_dim))
        self.pos_embed = self.sincos_pos_embed()
        
        self.decoder_embed = nn.Linear(emb_dim, dec_emb_dim, bias=True)    
        self.blocks = nn.ModuleList([
            Block(dec_emb_dim, num_head, mlp_ratio, qkv_bias=True, norm_layer=torch.nn.LayerNorm)
            for i in range(num_layer)])
        self.transformer = torch.nn.Sequential(*[Block(emb_dim, num_head) for _ in range(num_layer)])
        self.mask_token = nn.Parameter(torch.zeros(1, 1, dec_emb_dim))
        self.norm = torch.nn.LayerNorm(dec_emb_dim)
        self.decoder_pred = nn.Linear(dec_emb_dim, patch_size**2 * in_chans, bias=True)
        self.init_weight()
        
    def unpatchify(self,x):
        patch_embed = PatchEmbed(img_size, patch_size, in_chans, embed_dim)
        num_patches = patch_embed.num_patches
        p = patch_embed.patch_size[0]
        h = w = int(x.shape[1]**.5)
        assert h * w == x.shape[1]

        x = x.reshape(shape=(x.shape[0], h, w, p, p, 3))
        x = torch.einsum('nhwpqc->nchpwq', x)
        img = x.reshape(shape=(x.shape[0], 3, h * p, h * p))
        return img

    def sincos_pos_embed(self,max_sequence_length = 197, embedding_dim=512):

        position = torch.arange(0, max_sequence_length).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, embedding_dim, 2) * -(math.log(10000.0) / embedding_dim))
        position_encoding = torch.zeros(1, max_sequence_length, embedding_dim)
        position_encoding[0, :, 0::2] = torch.sin(position * div_term)
        position_encoding[0, :, 1::2] = torch.cos(position * div_term)

        return position_encoding
    
    def init_weight(self):
        trunc_normal_(self.cls_token, std=.02)
        trunc_normal_(self.pos_embed, std=.02)

    def forward(self, x, ids_restore):
        
        x = self.decoder_embed(x)
        mask_tokens = self.mask_token.repeat(x.shape[0], ids_restore.shape[1] + 1 - x.shape[1], 1)
        x_ = torch.cat([x[:, 1:, :], mask_tokens], dim=1)  
        x_ = torch.gather(x_, dim=1, index=ids_restore.unsqueeze(-1).repeat(1, 1, x.shape[2]))  
        x = torch.cat([x[:, :1, :], x_], dim=1) 
        x = x + self.pos_embed
        for blk in self.blocks:
            x = blk(x)
        x = self.norm(x)
        x = self.decoder_pred(x)
        x = x[:, 1:, :]
        x = self.unpatchify(x).squeeze(0)
        x = torch.einsum('chw->hwc', x)
        
        return x

decoder = v_Decoder()
x, _, ids_restore = encoder(img)
pred_img = decoder(x,ids_restore)
pred_img.shape
```
